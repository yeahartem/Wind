```python
def GPD(name, tres_list = [20, 25, 30], tres_exp = [0.1, 0.2, 0.3], tres_cox = [[20, 25, 30], [200, 300, 400], [2500, 5300, 8500], [50000, 100000, 200000]], lbda = [1, 2, 3, 4], exp = True, cox = True):
    # os.mkdir(name)
    data_max = Data.loc[Data['Название метеостанции'] == name]['Максимальная скорость']
    data_max  = data_max + [random.random() for _ in range(len(data_max))]
    data_max.dropna(inplace=True)
    data_max = data_max.ravel() 
    for tr in tres_list:
        thresh_modeling.gpdfit(data_max, tr, 'mle')
        thresh_modeling.return_value(data_max, tr, 0.05, 365, 36500, 'mle')
    # data_max = np.exp(data_max)
    # for tr in tres_exp:
    #     thresh_modeling.gpdfit(data_max, tr, 'mle')
    #     thresh_modeling.return_value(data_max, tr, 0.05, 365*8, 36500, 'mle')
    for l in lbda:
        data_max = Data.loc[Data['Название метеостанции'] == name]['Максимальная скорость']
        data_max.dropna(inplace=True)
        data_max  = data_max + [random.random() for _ in range(len(data_max))]
        data_max = st.boxcox(data_max, l)
        data_max = data_max.ravel()
        for tr in tres_cox[l-1]:
            thresh_modeling.gpdfit(data_max, tr, 'mle')
            thresh_modeling.return_value(data_max, tr, 0.05, 365, 36500, 'mle')
    
    
```
