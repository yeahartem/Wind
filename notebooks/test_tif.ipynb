```python
from osgeo import gdal
from matplotlib import pyplot as plt
import numpy as np
import os
from src.data_utils import data_processing as dp
```

# Read data

```python
path_to_data = 'data/history'
feature_name = 'tasmax'
file_paths = [path_to_data + '/' + fn for fn in os.listdir(path_to_data) if (fn[-4:] == '.tif' ) and feature_name in fn]
print("Numbe of .tifs: ", len(file_paths))
dataset = gdal.Open(file_paths[0], gdal.GA_ReadOnly) 
print("Number of rasters: ", dataset.RasterCount, "(days)")
print("Years: ", dataset.RasterCount / 365)
```
